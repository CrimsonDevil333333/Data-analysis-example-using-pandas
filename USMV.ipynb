```python
act = pd.DataFrame()
```

```python
for name,group in df.groupby("State"):
    if act.empty:
        act = group.set_index("Year")[["Low.2018"]].rename(columns={"Low.2018":name})
    else:
        act=act.join(group.set_index("Year")[["Low.2018"]].rename(columns={"Low.2018":name}))
```
